```python
f = Normal(4, 0.3)
size_sample = 250

sample = f.getSample(size_sample)
sample = np.array(sample).transpose()[0]

H_sample = np.array([h(x) for x in sample])

# Compute bin edges based on bin width
H_min = H_sample.min()
H_max = H_sample.max()

bin_width = (H_max - H_min)/30
bins = np.arange(H_min, H_max + bin_width, bin_width)

alpha = 0.95
q_H = np.quantile(H_sample, alpha)
average_H = np.average(H_sample)

# Plot vertical scaled histogram manually
#fig, ax = plt.subplots()

fig, ax = plt.subplots(1, 1, figsize=(9,6))

X = np.linspace(-2,15,300)

plt.style.use('seaborn-v0_8')
plt.grid('True')


ax.set_ylim(-0.2,2.6)
ax.set_xlim(-2,12)

# vertical line
ax.plot([0, 0], [-0.50, 3], color='black', lw = 1.5)
Zeros = np.zeros(len(sample))

# scatter plots
ax.scatter(sample, Zeros, color='blue', s=5, label='population sample')
ax.scatter(sample, [h(x) for x in sample], s=18, color = 'blue')
# ax
ax.plot(X, [h(x) for x in X], color='midnightblue', lw=3, label = 'objective function $G$')
ax.plot(X, [f.computePDF(x) for x in X], color='dodgerblue', lw=2.5, label = r'population density $f_\theta$')

ax.scatter(Zeros, H_sample, s=14, color = 'blue')
ax.hist([h(x) for x in sample], bins=bins, orientation='horizontal', 
         density=True, color='skyblue', label = r'density estimation of $Y|X\sim f_\theta$',
         edgecolor = 'black')
#ax.hist(sample, bins=200, 
#         density=True, color='darkred', label = r'density estimation of $f_\theta$',
#         edgecolor = 'black')

ax.scatter([0], [q_H], color = "orange", s = 60, label = r'alpha-quantile estimation of $Y|X\sim f_\theta$')
ax.scatter([0], [min_value_H], color = "green", s = 60, label = r'maximum value of $G$')

# labels
ax.set_xlabel(r"$X$", fontsize=15)
ax.set_ylabel(r"$Y$", fontsize=15)
ax.legend(fontsize=13, loc='upper right')


print(f"Estimated {int(alpha*100)}%-quantile of $G$ given $f$ is {q_H}")
print(f"True maximum of h: {min_value_H}")
print(f"Expected value of $h$ on the sample: {average_H}")


# ax[1].hist([h(x) for x in sample], bins=bins, 
#          density=True, color='skyblue', label = r'density estimation of $G(X)|X\sim f_\theta$',
#          edgecolor = 'black');

# ax[1].scatter([q_H], [0], color = "orange", s = 60, label = r'alpha-quantile estimation of $G(X)|X\sim f_\theta$')
# ax[1].scatter([min_value_H], [0], color = "green", s = 60, label = r'maximum value of $G$')
# ax[1].scatter([average_H], [0], color = 'darkred', s=60, label = r"sample average of $G$")
# ax[1].set_ylim(0,20)
# ax[1].set_xlim(1,2)
# ax[1].legend(fontsize = 10);
# ax[1].text(1.2, 10, f"q_95 = {q_H}\n G_max = {min_value_H}", fontsize=14)
fig.savefig("/home/bketema/Downloads/Figures_EYSM/sto_opt_4.pdf", dpi=600, bbox_inches='tight')
```

```python
def omega(alpha: float, array: np.ndarray):

    N = len(array)
    arange = np.arange(1, N+1)
    arg = (arange + np.ones(N)/2)/N
    omegas = arg >= alpha
    sorted_index = np.argsort(np.argsort(array))

    return omegas[sorted_index]/N

def utility(lamb, array):
    N = len(array)
    sorted_index = np.argsort(-array)
    arange = np.arange(1, N+1)
    num = [max(0, np.log(lamb/2 +1) - np.log(i)) for i in arange]
    den = np.sum(num)

    util = num/den - 1/lamb*np.ones(N)
    return util[np.argsort(sorted_index)]
```

```python

f0 = ExtendedTrNormal(0, 1, -1, 1)
eta = 10
nbIter = 50

polynomial_opt = RiemannianStochasticOptimization(h)
iterations, empMean, empQuant = polynomial_opt.naturalEvolutionStrategy(f0, eta=eta, nbIter=nbIter, n_k=50)

plt.figure(figsize=(14,6))
plt.grid('True')
colors = [(0.7, 0.2, p) for p in np.linspace(0,1, len(iterations), endpoint=True)]

X = np.linspace(-2,15,300)
plt.ylim(-0.2,2.5)
plt.xlim(-2,12)
plt.plot(X, [h(x) for x in X], color='dodgerblue', lw=3, label = 'objective function $h$')


for i, f in enumerate(iterations):
    plt.plot(X, [f.computePDF(x) for x in X], color=colors[i])
    #plt.scatter(samples[i], samples[i]*0, color = "darkred")

plt.plot(X, [iterations[0].computePDF(x) for x in X], color = "black")
plt.show()

plt.figure(figsize=(14,6))
arange = np.arange(1, nbIter+1)
plt.plot(arange, empMean, "o--",label="average")
plt.plot(arange, empQuant, "o--", label="quantile")
plt.plot(arange, np.ones(nbIter)*min_value_H, label="maximum value")
plt.ylim(-0.2, 2.2)
plt.legend(fontsize=12)
```
