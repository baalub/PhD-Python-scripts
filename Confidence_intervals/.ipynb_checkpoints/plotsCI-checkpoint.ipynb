```python
import openturns as ot
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy.optimize import minimize
from likelihood import Likelihood

# import modules from another directory
import sys
sys.path.append('/home/bketema/Python workspace/Multivariate case')

from truncatedDistribution import TruncatedDistribution  
from normal import Normal
from lognormal import LogNormal
```

```python
# define the input distributions
inputDoE = pd.read_csv("Bethsy_mc2_data/inputoutput.csv") # this is the list of input (vector) sample
outputVals = pd.read_csv("Bethsy_mc2_data/listMAX.csv")   # this will be the Y_sample

# renaming the columns (could have been done previously) and extracting the MC2
outputVals.rename(columns={'Unnamed: 0':'Evaluation number', '0':'MC2'}, inplace=True)

Y_sample = outputVals["MC2"]
X_sample = inputDoE["X38"]

n = len(Y_sample)
alpha = 0.5
beta = 0.95

# Defining the nominal distribution

lowerTrunc = 11
upperTrunc = 17

bounds = [(lowerTrunc, upperTrunc)]
interval = ot.Interval(lowerTrunc, upperTrunc)

m = (lowerTrunc + upperTrunc)/2   # m = 14
s = (upperTrunc - m)/3            # s = 1

f38 = TruncatedDistribution(Normal(m,s), interval)
```

```python
# 1) find a perturbation direction such that two-sided Bennett is better than Hoeffding

delta = 1
J = f38.fisherInformation()

angles = [np.pi/4, np.pi/2.1, 2*np.pi/3, 4*np.pi/3]
directions = []

for t in angles:
    v = np.array([np.cos(t), np.sin(t)])
    v = delta * v/np.sqrt(np.transpose(v)@J@v)
    directions.append(v)

perturbedDists = []

for v in directions:
    f_th = f38.exponentialMap(v)
    perturbedDists.append(f_th)


X = np.linspace(lowerTrunc-0.1, upperTrunc+0.1, 100)
plt.plot(X, [f38.computePDF(x) for x in X], color = 'green', label = "nominal")

for f_th in perturbedDists:
    plt.plot(X, [f_th.computePDF(x) for x in X], color = 'red', label = "perturbed")
```

```python
# 2) build the sequential exponential map

discretizeNum = 10
listDiscretizedGeodesics = []

for v in directions:
    discGeodesics = f38._sequentialExponentialMap(v, discretizeNum)
    listDiscretizedGeodesics.append(discGeodesics)
```

```python
quantEstim = []
bounds = []
variances = []

HCIlb = []
HCIub = []
BCIlb = []
BCIub = []

for geodesic in listDiscretizedGeodesics:
    quantsAlongGeodesics = []
    boundsAlongGeodesics = []
    varianceAlongGeodesics = []
    HCI_lb_AlongGeodesics = []
    HCI_ub_AlongGeodesics = []
    BCI_lb_AlongGeodesics = []
    BCI_ub_AlongGeodesics = []

    for f_th in geodesic:

        # estimate the quantile along the geodesic
        L = Likelihood(f_th, f38, pushforward=True)
        q = L.empiricalWeightedQuantile(Y_sample, alpha, X_sample)
        quantsAlongGeodesics.append(q)

        # compute the likelihood bounds
        tup = L.computeLikelihoodBounds(truncated=True, truncationBounds=[lowerTrunc, upperTrunc], method ="scipy")
        L.setLikelihoodBounds(tup)

        boundsAlongGeodesics.append(tup[1] - tup[0])
        varianceAlongGeodesics.append(tup[2])

        # compute the confidence intervals
        HoeffCI = L.computeCI(Y_sample, alpha, beta, X_sample = X_sample, CiBoundType="Hoeffding")
        BennCI = L.computeCI(Y_sample, alpha, beta, X_sample = X_sample, CiBoundType="Bennett")

        HCI_lb_AlongGeodesics.append(HoeffCI[0])
        HCI_ub_AlongGeodesics.append(HoeffCI[1])
        BCI_lb_AlongGeodesics.append(BennCI[0])
        BCI_ub_AlongGeodesics.append(BennCI[1])

    quantEstim.append(quantsAlongGeodesics)
    bounds.append(boundsAlongGeodesics)
    variances.append(varianceAlongGeodesics)

    HCIlb.append(HCI_lb_AlongGeodesics)
    HCIub.append(HCI_ub_AlongGeodesics)

    BCIlb.append(BCI_lb_AlongGeodesics)
    BCIub.append(BCI_ub_AlongGeodesics)
```

```python
Delta = []

for i in range(len(listDiscretizedGeodesics[0])):
    delta_inc = delta/discretizeNum *(i+1)
    Delta.append(delta_inc)

print(Delta)
```

```python
fig, ax = plt.subplots(4, 3, figsize=(24, 19))
#ax = ax.flatten()

fig.subplots_adjust(hspace=0.35)  # Increase vertical space

plt.style.use('seaborn-v0_8')

X = np.linspace(lowerTrunc-0.5, upperTrunc+0.5, 200)

for i, geodesic in enumerate(listDiscretizedGeodesics):
    
    for j, f_th in enumerate(geodesic):
        #plot the perturbed distributions
        ax[i][0].plot(X, [f_th.computePDF(x) for x in X], lw =1.8, color = "red", label='_nolenged_', alpha = 0.35)

    # plot the nominal distribution
    ax[i][0].plot(X, [f38.computePDF(x) for x in X], lw=2, color = "dodgerblue", label="nominal")

    # configure the necessary parameters for first column plots
    ax[i][0].plot([], [], lw=1.4, color = "red", alpha = 0.35, label = 'progressive perturbations') # dummy plot for legend
    ax[i][0].set_xlabel("$X_{38}$",fontsize="18")
    ax[i][0].set_ylabel("",fontsize="15")
    ax[i][0].tick_params(axis='both', labelsize=15) 
    ax[i][0].grid(True)
    ax[i][0].legend(fontsize = 14, loc='upper left')
    ax[i][0].set_title(f"({i+1}a)", fontsize=17)


    # plot quantile estimator
    ax[i][1].plot(Delta, quantEstim[i], color = "green", marker='o', linestyle='--', label = 'median estimation')

    #plot Hoeffding and Bennett CIs
    ax[i][1].plot(Delta, HCIlb[i], lw =1.5, marker='o', linestyle='--', color = "blue", label = f'{int(beta*100)}%-CI Hoeffding')
    ax[i][1].plot(Delta, HCIub[i], lw =1.5, marker='o', linestyle='--', color = "blue")
    ax[i][1].plot(Delta, BCIlb[i], lw =1.5, marker='o', linestyle='--', color = "red", label = f'{int(beta*100)}%-CI Bennett')
    ax[i][1].plot(Delta, BCIub[i], lw =1.5, marker='o', linestyle='--', color = "red")

    # shade the area between the interval bounds
    ax[i][1].fill_between(Delta, BCIlb[i], BCIub[i], color='red', alpha=0.1)
    ax[i][1].fill_between(Delta, HCIlb[i], HCIub[i], color='lightblue', alpha=0.3)

    # configure parameters
    ax[i][1].set_xlabel(r"$\delta$ (perturbation level)",fontsize="15")
    ax[i][1].set_ylabel("Temperature (C)",fontsize="15")
    ax[i][1].tick_params(axis='both', labelsize=15) 
    ax[i][1].grid(True)
    ax[i][1].legend(fontsize = 14, loc='upper left')
    ax[i][1].set_title(f"({i+1}b)", fontsize=17)
    if i == 3:
        ax[i][1].set_ylim(280 ,287.5)
        
                          
    # plot the bounds and variances on the likelihood
    ax[i][2].plot(Delta, variances[i], color = "green", marker='o', linestyle='--', label = 'likelihood variance')
    ax[i][2].plot(Delta, bounds[i], color = "purple", marker='o', linestyle='--', label = 'likelihood bound size ($b-a$)')
                                
    ax[i][2].tick_params(axis='both', labelsize=15)  
    ax[i][2].set_xlabel(r"$\delta$ (perturbation level)",fontsize="15")
    ax[i][2].set_ylim(0, 55)
    ax[i][2].set_title(f"({i+1}c)", fontsize=17)
    ax[i][2].grid(True)
    ax[i][2].legend(fontsize = 14, loc='upper left')
    if i==3:
        ax[i][2].set_ylim(0,6)

#plt.savefig("figures/H vs B median comparison.pdf", dpi=600, bbox_inches='tight')
```

## Equivalent perturbations of six different inputs

```python
from inputCode import SimulationCodeInputs
```

```python
# define the inputs i.e. variable names, numbers and type of distributions
inputNames = ["STOIGD", "STOIDO", "SPSVBR", "SPIVBR", "SPCV", "SCLFBR"]
inputVariables =["X4", "X7", "X18", "X32", "X34", "X38"]

distTypeStr ={"X4": "LogNormal", 
            "X7": "LogNormal", 
            "X18": "LogNormal", 
            "X32": "LogNormal", 
            "X34": "LogNormal",
            "X38": "Normal"}

# define the truncation bounds, note that the trim is always taken 3
truncationBounds = {"X4": [0.23, 3.5], 
                    "X7": [0.1, 10],
                    "X18": [0.236, 6.046],
                    "X32": [0.236, 6.046], 
                    "X34": [0.627, 1.259],
                    "X38": [11, 17]}

# define an instance from the class SimulationCodeInputs
inputs = SimulationCodeInputs(
    inputVariables=inputVariables, inputNames=inputNames, distTypeStr=distTypeStr, truncationBounds=truncationBounds,
    Dependence=False
)

# given the information above, we have a method buildInputDistribution which builds the input
# distribution
inputDist = inputs.buildInputDistribution()

# set the input distribution previously specified to the inputs object
inputs.setInputDistribution(inputDist)

# build the perturbed distributions for each inputs
amounts = 10 # number of perturbed distributions for each input
delta = 0.8 # the perturbation level in terms of the Fisher-Rao distance

dictPerturbedDistributions = inputs.buildPerturbedDistributions(amounts, FisherRao = True, delta=delta)
```

```python
# extract the sample from the inputs that will be perturbed and add to the dictionnary with the variable tag
X_sample = {}
for i in inputVariables:
    X_sample[i] = inputDoE[i][:n]

# and set the input sample to the inputs object
inputs.setInputSample(X_sample)
 
# compute quantile estimator and Wilks interval of 95% confidence
alpha = 0.95
empiricalQuantile = np.quantile(Y_sample, alpha)
print(f"95% empirical quantile = {empiricalQuantile}")
# compute Wilks interval: TODO
```

```python
# quantile order for the quantile estimation for perturbed distributions
alpha = 0.95

# compute the IS quantile estimator for each perturbed distributions of each inputs
dictEstimatedQuantiles = inputs.estimatePerturbedQuantiles(
    alpha=alpha, dictPerturbedDistributions=dictPerturbedDistributions, Y_sample=Y_sample
)

# get the maximum (i.e. worst case) quantile estimator for each input
dictWorstCaseDistributions = inputs.computeMaxEstimation(
    dictPerturbedDistributions, dictEstimatedQuantiles
)
```

```python
# for each input, create a subplot
n = len(inputVariables)
r = n%3
if r ==0:
    d = int(n/3)
else:
    d = int(n/3) +1

plt.style.use('seaborn-v0_8')
fig, ax = plt.subplots(d, 3, figsize=(16, 9))
ax = ax.flatten()
fig.subplots_adjust(hspace=1)  # Increase vertical space

# iterate on the number of perturbed inputs
for i, input in enumerate(inputVariables):
    lbT, ubT = inputs.truncationBounds[input][0], inputs.truncationBounds[input][1]
    f_i = inputDist.getMarginal(i) # nominal marginal distribution
    f_max = dictWorstCaseDistributions[input]

    e = (ubT -lbT)*0.08
    X = np.linspace(lbT-e, ubT+e, 500)
    # plot the nominal distribution
    ax[i].plot(X, [f_i.computePDF(x) for x in X], lw=2.5, color = "dodgerblue", label="nominal")
    name_str = inputVariables[i]
    ax[i].set_xlabel(xlabel=f"$X_{{{name_str[1:]}}}$", fontsize="18")

    for j in range(amounts):       
        # plot the perturbed distributions
        f_th = dictPerturbedDistributions[input][j]
        ax[i].plot(X, [f_th.computePDF(x) for x in X], lw =1.4, color = "red", label='_nolenged_', alpha = 0.25)
        
    # plot worst case distribution
    ax[i].plot(X, [f_max.computePDF(x) for x in X], lw =2.5, color = "darkred", label = "worst case")
    ax[i].tick_params(axis='x', labelsize=14)
    ax[i].tick_params(axis='y', labelsize=14)

    ax[i].set_title(f"{inputs.inputNames[i]}", fontsize=15)

    #dummy plot for legend
    ax[i].plot([], [], color = "red", label='perturbed', alpha = 0.3)            

    ax[i].legend(fontsize=16)
    ax[i].grid(True)

if r != 0:
    ax[int(d*3-1)].axis('off') 

fig.tight_layout()
fig.savefig("figures/equivalent perturbed distr delta=0.8.pdf", dpi=600, bbox_inches='tight')
```

```python
# build the N-A confidence intervals for the worst case distributions of each specified input

# we specify the parameters
alpha = 0.95 # quantile order
beta = 0.95 # confidence level
intervalType = "right bound"  # other options: "left bound" or "two-sided"
# epsilon is determined from alpha, beta and n

dictConfidenceIntervalsHoeff = inputs.quantileNACIs(
    dictWorstCaseDistributions, Y_sample, alpha, beta=beta, 
    intintervalType = intervalType, CiBoundType="Hoeffding"
)

dictConfidenceIntervalsBenn = inputs.quantileNACIs(
    dictWorstCaseDistributions, Y_sample, alpha, beta=beta, 
    intintervalType = intervalType, CiBoundType="Bennett"
)
```

```python
# compute worst case estimated quantiles

worstQauntiles= {}
i=0

for input in inputVariables:
    fmax = dictWorstCaseDistributions[input]
    Like = Likelihood(fmax, inputDist.getMarginal(i), pushforward=True)
    q = Like.empiricalWeightedQuantile(Y_sample, alpha, X_sample=X_sample[input])
    worstQauntiles[input] = q
    i+=1

def printQuantileEstimations(inputs, worstQauntiles):

    for i, input in enumerate(inputs.inputVariables):
        print(f"for {input} ({inputNames[i]}): {int(alpha*100)}%-quantile is estimated at {worstQauntiles[input]:.2f}C")
```

```python
#plot the empirical weighted quantile estimators as a function of the sample size

stepSize = 50
N = np.arange(1, 10000, stepSize)
nbIterations = int(10000/stepSize)

plt.figure(figsize=(16,9))
plt.style.use('seaborn-v0_8')
plt.tick_params(axis='x', labelsize=14)
plt.tick_params(axis='y', labelsize=14)


colors = ["red", "blue", "green", "teal", "purple", "orange"]
for j, input in enumerate(inputVariables):
    QuantileCurve = []

    fmax = dictWorstCaseDistributions[input]
    Like = Likelihood(fmax, inputDist.getMarginal(j), pushforward=True)

    for i in N:
        q = Like.empiricalWeightedQuantile(Y_sample[:i], alpha=0.95, X_sample=X_sample[input][:i])
        QuantileCurve.append(q)

    plt.plot(N, QuantileCurve, color = colors[j], label=f"$\\hat q_{{{alpha}}}$ for perturbation of {inputVariables[j]} ", ls = '--')

plt.legend(fontsize=15)
plt.ylim(310, 360)
plt.xlabel("Sample size $n$", fontsize=15)
plt.ylabel("Temperature (C)", fontsize=15)

fig.savefig("figures/worst estimated quantile curves.pdf", dpi=600, bbox_inches='tight')

```

```python
#print on the terminal the confidence intervals for the worst perturbed quantiles for each input
print(f"{alpha*100}% quantile estimation")
printQuantileEstimations(inputs, worstQauntiles)
print("")
print("Using Heoffding ")
inputs.printQuantileCIs(dictConfidenceIntervalsHoeff, alpha)
print("")
print("Using Bennett")
inputs.printQuantileCIs(dictConfidenceIntervalsBenn, alpha)
```
