# Non asymptotic confidence interval for importance sampling estimator of quantiles

Let $\mu$ and $\mu_0$ be two probability measures on $\mathbb R$ such that $\mu\ll \mu_0$ with a known likelihood ratio $L:=\frac{d\mu}{d\mu_0}$. Given a sample $Y_1,...,Y_n$ of $\mu_0$ we would like to estimate $q_\alpha(\mu)$ and build a confidence interval for the latter. 

We use the reverse importance sampling technique to build the following estimator:

$$\widehat q_{\alpha,is} = \inf\{t \in \mathbb R \ | \ \widehat{F_\mu}(t)\geq \alpha \}$$

where $\widehat{F_\mu}$ defined as

$$\widehat{F_\mu}(t) = \frac{1}{\sum_{j=1}^n L(Y_j)}\sum_{j=1}^n L (Y_j)\textbf{1}_{Y_j\leq t}$$

This quantile estimator, under some assumptions given in [Glynn, 1996], verifies a central limit theorem

$$\sqrt{n}\big(\widehat q_{\alpha,is} - q_{\alpha}\big) \longrightarrow \mathcal N(0,\sigma_\infty^2),$$

$\sigma_\infty^2=\frac{\mathbb E_{\mu_0} \big[L(Y)^2 (\textbf{1}_{Y\leq q_{\alpha}} - \alpha)^2\big]}{F_{\mu}'(q_{\alpha})^2}.$

This is not useable in practice. We then prove the following result for building confidence intervals

--------------------------------------------------------------

$\textbf{Proposition.}$ If the likelihood ratio $L$ is bounded by $\beta$, then for all $n\geq 1$ and $\xi>0$ we have
$$\mathbb P\Big(\widehat q_{\alpha-(\xi+\varepsilon),i\delta}\leq q_{\alpha,i\delta}\leq \widehat q_{\alpha+\xi+\varepsilon,i\delta}\Big) \geq 1- 2\sum_{k=1}^\gamma \exp\left(\frac{-2n\xi^2}{\beta^2\left(1+\frac{k}{\gamma+1}+\xi\right)^2}\right),$$
where $\gamma\geq 1$ is a whole number parameter and $\varepsilon=\frac{1}{\gamma+1}$.

----------------------------------------------

$\textbf{Remarks.}$ The bounds of the interval are elements of the initial sample $\{Y_1,...,Y_n\}\cup\{\pm \infty \}$ and the upper bound of the interval will be infinite if $\alpha+\xi+\varepsilon\geq 1$. Since usually $\alpha$ is close to one, this implies that $\xi+\varepsilon\leq 1-\alpha$. In addition, the confidence level is independent from $\alpha$, the infimum of the left hand side (i.e. the worst case) in the Proposition is still above the obtained confidence level.

Let us consider the case where $\mu=\mathcal N(0,1)$ and $\mu_0 = \text{Lap}(0,1)$, the likelihood ratio $L$
$$ L(x) = \sqrt{\frac{2}{\pi}}e^{-\frac{x^2}{2}+|x|}$$
is bounded by $\beta = \sqrt{\frac{2}{\pi}}e^{-\frac{1}{2}}$.

```python
import numpy as np
import matplotlib.pyplot as plt
import timeit
import scipy.stats as scs
```

```python
beta = np.sqrt(2/np.pi)*np.exp(-0.5)*3
print(beta)
```

```python
def confidence_level_Hoeff(gamma,n,xi):
    f = -2*n*xi**2/(beta**2 *(2 + xi)**2)
    g = -2*n*xi**2/(9*beta**2) 
    
    liste = [np.exp(-2*n*xi**2/(beta**2*(1+k/(gamma+1) + xi)**2) ) for k in range(1,gamma+2)]
    return 1-gamma*(np.exp(f) + np.exp(g)), 1 - 2*np.sum(liste)

def confidence_level_Hoeff_improved(gamma,k,n,xi):
    return 1 - np.exp(-2*n*xi**2/(beta**2*(1+k/(gamma+1) + xi)**2)) - np.exp(-2*n*xi**2/(beta**2*(1+(k+1)/(gamma+1) + xi)**2))
           
                      
def h(x):
    return (1+x)*np.log(1+x) - x

def confidence_level_Ben(gamma,n,xi):
    #b = beta**2 +beta*(1-2*xi)+2*xi*beta**2+xi**2*beta
    b = beta**2
    #b_tilde = 2*beta**2 + xi**2*beta**2 + 2*xi*beta**2
    b_tilde = beta**2*(1+xi)**2
    
    f = -n*b*h(beta*xi/b)/beta**2
    g = -n*b_tilde*h(beta*xi/b_tilde)/beta**2
    return 1-gamma*(np.exp(f) + np.exp(g))
```

```python
gamma = 100
eps = 1/(gamma+1)
n = 10000
xi = 0.01
k = 50   # k est la partie enti√®re de gamma*alpha

print(f"Improved Hoeffding ={confidence_level_Hoeff_improved(gamma,k,n,xi)}")
print(f"Hoeffding={confidence_level_Hoeff(gamma,n,xi)}")
print(f"Bennett={confidence_level_Ben(gamma,n,xi)}")                   
```

Let us first build the importance sampling based estimator of $F_\mu$

```python
def L(x):
    return np.sqrt(2/np.pi)*np.exp(-x**2/2 + np.abs(x))

T = np.linspace(-3,3,100)
plt.plot(T,L(T))
print(np.max(L(T)))
```

```python
def F_is(Y_sample,t):

    Likelihood = L(Y_sample)  
    G = np.array([Y_sample, ] * len(t))  
    T = np.array([t, ] * len(Y_sample))  

    Bool = G <= T.transpose()       

    M = np.array([Likelihood, ] * len(t))     
    B = M * Bool                
    
    return np.sum(B, axis=1) / np.sum(Likelihood)
```

```python
n = 1000
Y_sample = scs.laplace().rvs(n)
plt.plot(T,F_is(Y_sample,T))
```

We now build the importance sampling based quantile estimator

```python
def quantile_estim(Y_sample,alpha):

    Y_ord = np.sort(Y_sample)
    n = len(Y_sample)
    i = int(alpha*n)  # element of {1,...,n}
    #print(i)
    # if the cdf is smaller than alpha on the ith element of the ordered sample
    # we increment i
    if F_is(Y_sample,[Y_ord[i]]) < alpha:
        while F_is(Y_sample,[Y_ord[i]]) < alpha:
            i = i+1
            #print(i)
            if i >= n:
                return '+ inf'
        return Y_ord[i-1] # the -1 is because of python's indexing convention

    # if the cdf is larger than alpha on the ith element of the ordered sample 
    # we decrement i
    while F_is(Y_sample,[Y_ord[i]])>= alpha:
        i = i-1
        #print(i)
        if i < 0:
            return '- inf'
    return Y_ord[i] #again we use i instead of i-1 because of python's indexing convention
```

```python
n = 100
alpha = 0.8
Y_sample = scs.laplace().rvs(n)
#print(np.sort(Y_sample))
#print(F_is(Y_sample,np.sort(Y_sample)))
print(quantile_estim(Y_sample,alpha))
```

```python
def confidence_interval(Y_sample,alpha,xi,gamma):
    eps = 1/(gamma+1)
    
    lower = quantile_estim(Y_sample, alpha - xi - eps)
    print(lower)
    upper = quantile_estim(Y_sample, alpha + xi + eps)
    return np.array([lower, upper])
```

```python
n = 66000
Y_sample = scs.laplace().rvs(n)
alpha = 0.95

gamma = 100
xi = 0.005
print(alpha + xi + 1/(gamma+1))
print(alpha - xi - 1/(gamma+1))

print("confidence level =", confidence_level_Hoeff_improved(gamma,int(alpha*gamma),n,xi))
print("")
start = timeit.default_timer()
print("for the confidence interval =",confidence_interval(Y_sample,alpha,xi,gamma))
stop = timeit.default_timer()
print("")
print("true value =", scs.norm().ppf(alpha))
print(f"computation time = {stop-start} seconds")
```
