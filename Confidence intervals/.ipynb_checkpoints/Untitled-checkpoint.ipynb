# 1. Comparison between Wilks bound and Cramér-Chernoff bound in the standard case

```python
def cramer_bound(n, xi, a):
    
    A = ((1-a-xi)/(1-a) )**(-n*(1-a-xi))
    B = ((a+xi)/a)**(-n*(a+xi))
    C = ((1-a+xi)/(1-a) )**(-n*(1-a+xi))
    D = ((a-xi)/a)**(-n*(a-xi))
    
    return 1 - A*B - C*D

def wilks_bound(n, xi, a):
    i = np.floor(n*(a-xi))
    j = np.floor(n*(a+xi))
    
    return sc.special.betainc(i,n-i+1,a) - sc.special.betainc(j,n-j+1,a)
   
n = 100
xi = 0.01
alpha = 0.95
print(cramer_bound(n,xi,alpha))
print(wilks_bound(n,xi,alpha))
```

```python
n_line = np.linspace(0,5000,200)
Wilks = [wilks_bound(n,xi,alpha) for n in n_line]
Cramer = [cramer_bound(n,xi,alpha) for n in n_line]

plt.figure(figsize=(12,8))
plt.xlabel('Sample size n',fontsize="12")
plt.ylabel("Confidence level",fontsize="12")
plt.plot(n_line,Wilks, label='Exact probability $p_n$')
plt.plot(n_line,Cramer, label = 'Cramér-Chernoff bound', color ="red")
plt.legend(fontsize="12");
```

```python
Alpha = np.linspace(0.01,0.99,100)

plt.figure(figsize=(12,8))
plt.xlabel('Sample size n',fontsize="12")
plt.ylabel("Quantile order $\alpha$",fontsize="12")
plt.plot(Alpha, [wilks_bound(n= 5000,xi= 0.01, a= A) for A in Alpha],label='Exact probability $p_n$')
plt.plot(Alpha, [cramer_bound(n= 5000,xi= 0.01, a= A) for A in Alpha], color ="red", label = 'Cramér-Chernoff bound')

plt.legend(fontsize="12");
```

We will build confidence intervals for quantiles in the importance sampling case where $\mu_0=$ Laplace and $\mu=\mathcal N(0,1)$. The likelihood ratio is given by
$$L(x) = \frac{d\mu}{d\mu_0} = \frac{(2\pi)^{-1/2}e^{-x^2/2}}{1/2e^{-|x|}}=\sqrt{\frac{2}{\pi}}e^{-x^2/2 + |x|}$$

This likelihood ratio is bounded from above and the best bound is given by
$$\beta = L(1) = \sqrt{\frac{2}{\pi}}e^{-1/2 + 1}=\sqrt{\frac{2}{\pi}}e^{1/2}.$$
Moreover, there is no positive lower bound for $L$ since $\lim_{x\to\pm \infty}L(x) = 0$.

```python
def L_ratio(f,g,x):
    return f.pdf(x)/g.pdf(x)

def sample(f,n):
    return f.rvs(n)

def F_emp_L(f,g,Y_sample,t):
    n = len(Y_sample)
    L = np.array([f.pdf(x)/g.pdf(x) for x in Y_sample]) 
    T = L<=t
    return np.sum(T)/n

def F_is(f,g,Y_sample,t):

    Likelihood = f.pdf(Y_sample)/g.pdf(Y_sample)  
    G = np.array([Y_sample, ] * len(t))  
    T = np.array([t, ] * len(Y_sample))  

    Bool = G <= T.transpose()       

    M = np.array([Likelihood, ] * len(t))     
    B = M * Bool                
    
    return np.sum(B, axis=1) / np.sum(Likelihood)

def quantile_estim(f,g,Y_sample,alpha):

    Y_ord = np.sort(Y_sample)
    n = len(Y_sample)
    i = int(alpha*n)  # element of {1,...,n}
    # if the cdf is smaller than alpha on the ith element of the ordered sample
    # we increment i
    if F_is(f,g,Y_sample,[Y_ord[i]]) < alpha:
        while F_is(f,g,Y_sample,[Y_ord[i]]) < alpha:
            i = i+1
            if i >= n:
                return '+ inf'
        return Y_ord[i-1] # the -1 is because of python's indexing convention

    # if the cdf is larger than alpha on the ith element of the ordered sample 
    # we decrement i
    while F_is(f,g,Y_sample,[Y_ord[i]])>= alpha:
        i = i-1
        if i < 0:
            return '- inf'
    return Y_ord[i] #again we use i 

def confidence_interval(f,g,Y_sample,alpha,xi):

    lower = quantile_estim(f,g,Y_sample, alpha - xi)
    upper = quantile_estim(f,g,Y_sample, alpha + xi)
    
    return np.array([lower, upper])
```

```python
def conf_level_Hoeffding(n, xi, beta, alpha):
    c_plus = (1-alpha+xi)**2*beta/8
    c_minus = (alpha+xi)**2*beta/8
    
    e_plus = np.exp(-n*xi**2 / c_plus)
    e_minus = np.exp(-n*xi**2 / c_minus)
    
    return 1 - e_plus - e_minus
```

```python
f = scs.lognorm(1)
g = scs.lognorm(1)

n = 100
alpha = 0.95
xi = 0.02
Y_sample = f.rvs(n)

I = confidence_interval(f, g, Y_sample, alpha, xi)
print(I)
print(conf_level_Hoeffding(n,xi,1,alpha))
```

```python
def conf_level_Hoeffding(n, xi, beta, alpha):
    c_plus = (1-alpha+xi)**2*beta/8
    c_minus = (alpha+xi)**2*beta/8
    
    e_plus = np.exp(-n*xi**2 / c_plus)
    e_minus = np.exp(-n*xi**2 / c_minus)
    
    return 1 - e_plus - e_minus

def h(u):
    return (1+u)*np.log(1+u) - u


def conf_level_Bennett(n, xi, beta, alpha):
    # beta is the bound on the likelihood ratio, same as in Hoeffding
    b = beta 
    a = alpha

    b_pl = b*(a-xi)
    b_mi = b*(1-a-xi)
    
    nu_pl = max(0,b*a*(1-2*a+2*xi))+vL*(-a+xi)**2
    nu_mi = max(0,b*a*(1-2*a-2*xi))+vL*(-a-xi)**2
    
    e_pl = np.exp(-n*nu_pl/b_pl**2 *h(b_pl*xi/nu_pl) )
    e_mi = np.exp(-n*nu_mi/b_mi**2 *h(b_pl*xi/nu_mi) )
    return 1 - e_pl - e_mi

alpha = 0.5
n = 1000
xi = 0.02

print(conf_level_Hoeffding(n, xi, beta, alpha))
print(conf_level_Bennett(n, xi, beta, alpha))
```
