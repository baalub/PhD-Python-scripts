```python
import CI_Bethsy as bth
f_th = bth.perturbedDistributions_33[1]
f0 = bth.f_33
```

```python
def l(x):
    return f_th.computePDF(x)/f0.computePDF(x)

def l_square(x):
    return f_th.computePDF(x)**2/f0.computePDF(x)

lb, ub = bth.lowerTrunc_33, bth.upperTrunc_33

print(scp.integrate.quad(lambda x: f_th.computePDF(x)**2/f0.computePDF(x),
                              lb, ub))

x0 = (lb+ub)/2
print(x0)

minimize(lambda x: f_th.computePDF(x)/f0.computePDF(x), x0=x0, 
         bounds= [(lb, ub)]).fun
```
