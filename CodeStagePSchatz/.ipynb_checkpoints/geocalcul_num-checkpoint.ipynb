```python
def formal_fisher(dens, param, d=2):
    ldens=sp.expand_log(sp.log(dens),force=True)
    display(ldens)
    def info(i,j):
        der=sp.diff(ldens,param[i],param[j])
        integ=der*dens
        res=-sp.integrate(integ,(x1,-sp.oo,sp.oo),(x2,-sp.oo,sp.oo)) #v.a.r. à densité
        #res2=sp.integrate(dens2,(x1,-sp.oo,sp.oo),(x2,-sp.oo,sp.oo))
        return sp.simplify(res)
    I=sp.Matrix(d,d,info) #matrice évaluant info en chaque coordonnée
    #I=info(0,0)
    return I
```

```python
def fisher_distrib(distrib,N=10000):
    
    valparam=distrib.getParameter()
    d=len(valparam)
    n=distrib.getDimension()
    
    
    

    def f(param,x):
        distrib.setParameter(param)
        out=distrib.computePDF(x)
        distrib.setParameter(valparam)
        return out
    
    def quantile(param,u):
        
        v=np.array(u)
        distrib.setParameter(param)
        q=[]
        for i in range(n):
            
            margi=distrib.getMarginal(i)
            print(v)
            qi=margi.computeQuantile(v[i])[0]
            q.append(qi)
        distrib.setParameter(valparam)
        return ot.Point(q)
        

    
    print(distrib.computeLogPDFGradient(quantile(valparam,0.5)))
    
    
    def pythintegrand(u):
        
        q=quantile(valparam,u)
        
        gradlog=np.array(distrib.computeLogPDFGradient(q))[np.newaxis]
        pgradlog=gradlog*gradlog.T
        flatpgradlog=pgradlog.flatten()
        return flatpgradlog
    
    otintegrand=ot.PythonFunction(n,d**2,pythintegrand)

    
    experimentdistri=[]
    for i in range(n):
        experimentdistri.append(ot.Uniform())
    experimentMC = ot.MonteCarloExperiment(ot.JointDistribution(experimentdistri),N)
    sobol=ot.SobolSequence(n)
    copula=distrib.getCopula()
    experimentSobol=ot.LowDiscrepancyExperiment(sobol,copula,N)
    integration = otexp.ExperimentIntegration(experimentSobol)

    value = integration.integrate(otintegrand)    

    I_F=np.reshape(value,(-1,d))
    return I_F
    
```

```python
def fisher_gen(distrib,h=0.001,N=10000):
   #on suppose que les informations de Fisher de dens1 et dens2 sont calculables/connues
    n=distrib.getDimension()
    d=len(distrib.getParameter())
    marg=[distrib.getMarginal(i) for i in range(n)]
    cop=distrib.getCopula()
    param=[marg[i].getParameter() for i in range(n)]
    param.append(cop.getParameter())
    valuesparam=[]
    for i in range(n+1):
        valuesparam+=param[i]
    print(marg)
    
    def repart_marg(x,valparam,i):
        margi=marg[i]
        
        margi.setParameter(valparam)
        return margi.computeCDF(x)
    def quantile_marg(u):
        out = [marg[i].computeQuantile(u[i]) for i in range(n)]
        return out
    
    fisher_marg = [fisher_distrib(marg[i], N) for i in range(n)]
    fisher_marg.append(fisher_distrib(cop, N))
  
         
    d_vect=[0]
    for i in range(n+1):
        d_vect.append(d_vect[-1]+len(param[i]))


    

    
    

    valuescop=cop.getParameter()
    dcop=len(param[-1])

    
    
    pdfcop=cop.getPDF()
    def denscop(*u):
        return pdfcop(np.array(u))[0]
    def logpdfcop(u,valparam):
        
        cop_=cop
        cop_.setParameter(valparam)
        return cop_.computeLogPDF(u)
    def gradlogpdfcop(u,valparam):
        cop_=cop
        cop_.setParameter(valparam)
        return cop_.computeLogPDFGradient(u)
        


    
    
  
    Fisher=fisher_marg[0]
    for i in range(n):
        Fisher=scp.linalg.block_diag(Fisher,fisher_marg[i+1])

    def derivat(x,valparam):
        arg=ot.Point()
        
        values_l=[valparam[d_vect[i]:d_vect[i+1]] for i in range(n+1)]
        
        
        for i in range(n):
            arg.add(repart_marg(ot.Point(x[i]),values_l[i],i))
        
        return logpdfcop(arg,values_l[-1])
    delta=h/2*np.identity(d)
    
    def hesslog(u):
        q=quantile_marg(u)
        K=np.zeros((d,d))
        for i in range(d-dcop):
            for j in range(i,d):
                der=(derivat(q,valuesparam+delta[i]+delta[j])+derivat(q,valuesparam-delta[i]-delta[j])-derivat(q,valuesparam+delta[i]-delta[j])-derivat(q,valuesparam-delta[i]+delta[j]))/(h**2)
                K[i][j]=der
        Kt=K.T
        K_=K+Kt-K.diagonal()
        flatK=-K_.flatten()
        return flatK
    
    othesslog=ot.PythonFunction(n,d**2,hesslog)
    
    experimentdistri=[]
    for i in range(n):
        experimentdistri.append(ot.Uniform())
    experimentMC = ot.MonteCarloExperiment(ot.JointDistribution(experimentdistri,cop),N)
    sobol=ot.SobolSequence(n)
    
    experimentSobol=ot.LowDiscrepancyExperiment(sobol,cop,N)
    integration = otexp.ExperimentIntegration(experimentSobol)

    value = integration.integrate(othesslog)    
    value=np.array(value)
    value=np.reshape(value,(-1,d))
    Fisher+= value
    return Fisher
    
    
    
```

```python
def fstorderchristoffel_distrib(distrib,h=0.001,N=10000):
    valparam=distrib.getParameter()
    d=len(valparam)
    n=distrib.getDimension()
        
    delta=h/2*np.identity(d)
   
    def pythintegrand(u):
        
        
        q=ot.Point(u)
        gradlog=np.array(distrib.computeLogPDFGradient(q))[np.newaxis][np.newaxis]
        hesslog=[]
        
        for i in range(len(valparam)):
            distrib.setParameter(valparam+delta[i])
            
            up=np.array(distrib.computeLogPDFGradient(q))
            distrib.setParameter(valparam-delta[i])
            
            down=np.array(distrib.computeLogPDFGradient(q))
            der=up-down
            hesslog.append(der/h)
        print(hesslog)
        hesslog=np.array(hesslog)
        hesslog=hesslog[np.newaxis]
        T2gradlog=np.transpose(gradlog,[1,2,0])
        pgradlog=gradlog*T2gradlog
        
        s=hesslog+pgradlog/2
        Tgradlog=np.transpose(gradlog)
        out=s*Tgradlog
        
        flatout=out.flatten()
        
        return flatout
   
    print(pythintegrand([0.5,0.5]))
    
    otintegrand=ot.PythonFunction(n,d*d*d,pythintegrand)
    distrib.setParameter(valparam)
    copula=distrib.getCopula()
    
    ot.RandomGenerator.SetSeed(0)
    
    #experimentMC = ot.MonteCarloExperiment(copula,N)
    sobol=ot.SobolSequence(n)
    
    experimentSobol=ot.LowDiscrepancyExperiment(sobol,distrib,N)
    integration = otexp.ExperimentIntegration(experimentSobol)
    


    value = integration.integrate(otintegrand)    

    Christo1=np.reshape(value,(d,d,d))
    return Christo1
```

```python
def fstorderchristoffel_gen(distrib,h=0.001,N=10000):
    n=distrib.getDimension()
    d=len(distrib.getParameter())
    marg=[distrib.getMarginal(i) for i in range(n)]
    cop=distrib.getCopula()
    param=[marg[i].getParameter() for i in range(n)]
    param.append(cop.getParameter())
    valuesparam=[]
    for i in range(n+1):
        valuesparam+=param[i]
    valuesparam=distrib.getParameter()
    print(marg)
    
    def repart_marg(x,valparam,i):
        margi=marg[i]
        
        margi.setParameter(valparam)
        return margi.computeCDF(x)
    def quantile_marg(u):
        out=[marg[i].computeQuantile(u[i]) for i in range(n)]
        return out
    
    fisher_marg=[fisher_distrib(marg[i],N) for i in range(n)]
    fisher_marg.append(fisher_distrib(cop,N))
  
         
    d_vect=[0]
    for i in range(n+1):
        d_vect.append(d_vect[-1]+len(param[i]))


    

    
    

    valuescop=cop.getParameter()
    dcop=len(param[-1])

    
    
    pdfcop=cop.getPDF()
    def denscop(*u):
        return pdfcop(np.array(u))[0]
    def logpdfcop(u,valparam):
        cop.setParameter(valparam)
        out=cop.computeLogPDF(u)
        cop.setParameter(param[-1])
        return out
    def gradlogpdfcop(u,valparam):
        cop_=cop
        cop_.setParameter(valparam)
        return cop_.computeLogPDFGradient(u)
        


    
    christo_marg=[fstorderchristoffel_distrib(marg[i],h,N) for i in range(n)]
    christo_marg.append(fstorderchristoffel_distrib(cop,h,N))
  
    Christo=christo_marg[0]
    for i in range(n):
        Fisher=scp.linalg.block_diag(Fisher,fisher_marg[i+1])

    def derivat(x,valparam):
        arg=ot.Point()
        
        values_l=[valparam[d_vect[i]:d_vect[i+1]] for i in range(n+1)]
        
        
        for i in range(n):
            arg.add(repart_marg(ot.Point(x[i]),values_l[i],i))
        
        return logpdfcop(arg,values_l[-1])
    delta=h/2*np.identity(d)
    
    def integrand(u):
        q=quantile_marg(u)
        J=np.zeros(d,d,d)
        grad=distrib.computeLogPDGradient(q)
        for i in range(d):
            for j in range(d):
                for l in range(d):
                    jerk=(derivat(q,valuesparam+delta[i]+delta[j]+delta[l])+derivat(q,valuesparam+delta[i]-delta[j]-delta[l])+derivat(q,valuesparam-delta[i]-delta[j]+delta[l])+derivat(q,valuesparam-delta[i]+delta[j]-delta[l])-derivat(q,valuesparam+delta[i]+delta[j]-delta[l])-derivat(q,valuesparam-delta[i]+delta[j]+delta[l])-derivat(q,valuesparam+delta[i]-delta[j]+delta[l])-derivat(q,valuesparam-delta[i]-delta[j]-delta[l]))/(h**3)
                    hess=(derivat(q,valuesparam+delta[i]+delta[j])+derivat(q,valuesparam-delta[i]-delta[j])-derivat(q,valuesparam+delta[i]-delta[j])-derivat(q,valuesparam-delta[i]+delta[j]))/(h**2)
                    
                    J[i][j][l]=jerk+hess*grad[l]
        J1=-np.array(J)
        J2=np.transpose(J1)
        J3=np.transpose(J1,[1,2,0])
        J_=J1+J2-J3
        flatJ=J.flatten()
        return flatJ
    
    
    
    otintegrand=ot.PythonFunction(n,d**3,integrand)
    
    experimentdistri=[]
    for i in range(n):
        experimentdistri.append(ot.Uniform())
    experimentMC = ot.MonteCarloExperiment(ot.JointDistribution(experimentdistri,cop),N)
    sobol=ot.SobolSequence(n)
    
    experimentSobol=ot.LowDiscrepancyExperiment(sobol,cop,N)
    integration = otexp.ExperimentIntegration(experimentSobol)

    value = integration.integrate(otintegrand)    
    value=np.array(value)
    value=np.reshape(value,(d,d,d))
    Christo+= value
    return Christo
```
